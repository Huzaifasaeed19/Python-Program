```python
first_name=input("Enter the First name of the user: ")
last_name=input("Enter the Last name of the user: ")
print("Before Swapping: ",first_name,last_name)
temp=first_name
first_name=last_name
last_name=temp
print("After Swapping: ",first_name,last_name)
```
